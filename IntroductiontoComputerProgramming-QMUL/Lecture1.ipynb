```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xvals = np.arange(0, 2*math.pi, 0.01) # define x values for plotting
yvals = np.sin(xvals) # generate corresponding y values
plt.plot(xvals, yvals)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
xvals = np.arange(0, 2*math.pi, 0.01) # define x values for plotting
yvals = np.sin(xvals) # generate corresponding y values
plt.plot(xvals, yvals)
yvals2 = np.cos(xvals) # generate corresponding y values
plt.plot(xvals, yvals2)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### Plotting $z=f(x,y)$

```python
def f(x,y):
    return (1 - (x/2) + x ** 5 + y ** 3)*np.exp(- x ** 2 - y ** 2)
```

```python
# create an (x, y) grid
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

3d plot of $z = f(x,y)$

```python
ax = plt.axes(projection = "3d")
ax.plot_surface(X, Y, Z)
plt.show()
```

```python
plt.contourf(X, Y, Z)
plt.show()
```
